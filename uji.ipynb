```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Load dataset
df = pd.read_csv("sleep_health_lifestyle_dataset.csv")

# Bersihkan nama kolom agar lebih mudah diakses
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
df.rename(columns={"quality_of_sleep_(scale:_1-10)": "quality_of_sleep"}, inplace=True)

# Lakukan uji ANOVA
anova_model = ols('quality_of_sleep ~ C(bmi_category)', data=df).fit()
anova_table = sm.stats.anova_lm(anova_model, typ=2)

# Ekstrak p-value
anova_p_value = anova_table["PR(>F)"][0]

# Tampilkan hasil ANOVA
print("\n=== Hasil Uji ANOVA ===")
print(anova_table)
print(f"\nP-Value: {anova_p_value:.5f}")

# # Simpan hasil ANOVA ke CSV
# anova_table.to_csv("anova_results.csv")
# print("\nHasil ANOVA telah disimpan sebagai 'anova_results.csv'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("sleep_health_lifestyle_dataset.csv")

# Bersihkan nama kolom agar lebih mudah diakses
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
df.rename(columns={"quality_of_sleep_(scale:_1-10)": "quality_of_sleep"}, inplace=True)

# Buat boxplot untuk melihat distribusi kualitas tidur berdasarkan kategori BMI
plt.figure(figsize=(5, 3))
sns.boxplot(x=df['bmi_category'], y=df['quality_of_sleep'], palette="Set2")

# Tambahkan judul dan label
plt.title("Distribusi Kualitas Tidur Berdasarkan Kategori BMI", fontsize=14)
plt.xlabel("Kategori BMI", fontsize=12)
plt.ylabel("Kualitas Tidur (Skala 1-10)", fontsize=12)

# Tampilkan grafik
plt.show()
```
